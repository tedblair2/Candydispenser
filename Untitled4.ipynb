```python
from tkinter import *
from tkinter.messagebox import showerror, showinfo


class Dispenser:

  def __init__(self, window: Tk):
    self.window = window
    self.candy_stack = []
    self.max_size = 7

    # Spring
    self.spring_left = 220
    self.spring_right = 270
    self.spring_top = 100
    self.spring_bottom = 400
    self.spring_mvt = 25  
    self.a_y = self.spring_top
    self.b_y = 175
    self.c_y = 250
    self.d_y = 325
    self.e_y = self.spring_bottom

    #block 
    self.bar_bottom = self.spring_top + 23 
    self.bar_cx = (self.spring_left + self.spring_right) / 2
    self.bar_cy = (self.spring_top + self.bar_bottom) / 2 
    
    #canvas
    self.stack_canvas = Canvas(self.window, width=window_width, height=window_height,bg="white",
                                     relief=RAISED, bd=10)
    self.stack_canvas.pack(side=LEFT)
    
     #heading
    self.heading_name = Label(self.stack_canvas, text="Candy Dispenser", bg="white", fg="black",
                                    font=("Helvetica",20,"bold","italic"))
    self.heading_name.place(x=140,y=20)
    
    #create lines
    self.left = self.stack_canvas.create_line(220,400,220,95,width=4,smooth=True)
    self.right = self.stack_canvas.create_line(270,400,270,95,width=4,smooth=True)
    self.stack_canvas.create_line(self.spring_left, self.e_y, self.spring_right, self.e_y, width=4,
      smooth=True)

    # Draw spring
    self.a = self.stack_canvas.create_line(self.spring_left, self.a_y, self.spring_right, self.a_y, width=4,smooth=True)
    self.a_b = self.stack_canvas.create_line(self.spring_left, self.b_y, self.spring_right, self.a_y, width=4,smooth=True)
    self.b = self.stack_canvas.create_line(self.spring_left, self.b_y, self.spring_right, self.b_y, width=4,smooth=True)
    self.b_c = self.stack_canvas.create_line(self.spring_left, self.c_y, self.spring_right, self.b_y, width=4,smooth=True)
    self.c = self.stack_canvas.create_line(self.spring_left, self.c_y, self.spring_right, self.c_y, width=4,smooth=True)
    self.c_d = self.stack_canvas.create_line(self.spring_left, self.d_y, self.spring_right, self.c_y, width=4,smooth=True)
    self.d = self.stack_canvas.create_line(self.spring_left, self.d_y, self.spring_right, self.d_y, width=4,smooth=True)
    self.d_e = self.stack_canvas.create_line(self.spring_left, self.e_y, self.spring_right, self.d_y, width=4,smooth=True)
    self.stack_canvas.create_line(self.spring_left, self.e_y, self.spring_right, self.e_y, width=4,
      smooth=True)
    
    #Buttons
    Button(self.window,text="PUSH",fg="blue",bg="black",font=("Arial",15,"bold"),
        relief=RAISED,bd=7,command=self.push).place(x=90,y=535)
    Button(self.window, text="POP", fg="blue", bg="black", font=("Arial", 15, "bold"),
        relief=RAISED,bd=7,command=self.pop).place(x=360, y=535)
    Button(self.window, text="SIZE" ,fg="blue", bg="black", font=("Arial", 15, "bold"),relief=RAISED,
        bd=7,command=self.report_size).place(x=360, y=475)
    Button(self.window, text="TOP" ,fg="blue", bg="black", font=("Arial", 15, "bold"),relief=RAISED,
        bd=7,command=self.top).place(x=90, y=475)
    Button(self.window, text="IS EMPTY" ,fg="blue", bg="black", font=("Arial", 15, "bold"),relief=RAISED,
        bd=7,command=self.is_empty).place(x=200, y=535)

  def pop(self):
    if self.size() > 0:

      candy = self.candy_stack.pop()
      self.stack_canvas.delete(candy['bar'])
      self.stack_canvas.delete(candy['label'])

      self.update_dispenser('pop')
    else:
      showerror("Error","Dispenser Empty!")

  def push(self):
    if self.size() < self.max_size:
      self.candy_stack.append(self.draw_candy())  # Add candy to stack.
      self.update_dispenser('push')
    else:
      showerror("Error","Dispenser Full!")

  def draw_candy(self):
    bar = self.stack_canvas.create_rectangle(
      self.spring_left, self.spring_top, self.spring_right, self.bar_bottom,
      fill="black", width=2, outline="blue"
    )
    tag = f'C{self.size() + 1}'
    label = self.stack_canvas.create_text(
      self.bar_cx, self.bar_cy, text=tag, fill='white'
    )
    return {
      'bar': bar,
      'label': label,
      'tag': tag
    }


  def update_spring(self):#update spring
    self.stack_canvas.coords(self.a, self.spring_left, self.a_y, self.spring_right, self.a_y)
    self.stack_canvas.coords(self.a_b, self.spring_left, self.b_y, self.spring_right, self.a_y)
    self.stack_canvas.coords(self.b, self.spring_left, self.b_y, self.spring_right, self.b_y)
    self.stack_canvas.coords(self.b_c, self.spring_left, self.c_y, self.spring_right, self.b_y)
    self.stack_canvas.coords(self.c, self.spring_left, self.c_y, self.spring_right, self.c_y)
    self.stack_canvas.coords(self.c_d, self.spring_left, self.d_y, self.spring_right, self.c_y)
    self.stack_canvas.coords(self.d, self.spring_left, self.d_y, self.spring_right, self.d_y)
    self.stack_canvas.coords(self.d_e, self.spring_left, self.e_y, self.spring_right, self.d_y)

    self.stack_canvas.update()
        

  def update_dispenser(self, mode):
    # Update position of all candies excluding the topmost in the stack.
    if mode == 'push':
      for i in range(self.size()):
        self.update_candy_pos(self.candy_stack[i], (self.size() - 1) - i)

      # Update spring's pitch.
      self.a_y += 25 
      self.b_y += 20 
      self.c_y += 15
      self.d_y += 8
      self.update_spring()
    elif mode == 'pop':
      stack_size = self.size()
      for i in range(stack_size):
        self.update_candy_pos(self.candy_stack[i], stack_size - (i + 1))

      # Update spring's pitch.
      self.a_y -= 25 
      self.b_y -= 20
      self.c_y -= 15
      self.d_y -= 8
      self.update_spring()
    else:
      raise Exception
   

  def update_candy_pos(self, candy, y):
    updated_bar_top = self.spring_top + (self.spring_mvt * y)
    updated_bar_bottom = self.bar_bottom + (self.spring_mvt * y)

    self.stack_canvas.coords(
      candy['bar'], self.spring_left, updated_bar_top, self.spring_right, updated_bar_bottom
    )
    self.stack_canvas.coords(
      candy['label'], self.bar_cx, (updated_bar_top + updated_bar_bottom) / 2
    )

  def size(self):
    return len(self.candy_stack)

  def report_size(self):
    if self.size() == 0:
      showerror('Error', 'Dispenser Empty!')
    else:
      showinfo('Size', f'Number of candies is {self.size()}')

  def top(self):
    if self.size() == 0:
      showerror('Error', 'Dispenser Empty!')
    else:
      showinfo('Top', f'Top candy is {self.candy_stack[-1]["tag"]}')

  def is_empty(self):
    if self.size() == 0:
      showinfo('Empty', f'True')
    else:
      showinfo('Empty', f'False')


if __name__ == '__main__':
  window_height = 600
  window_width = 550

  root = Tk()
  root.title('Candy Dispenser')
  root.maxsize(window_width, window_height)
  root.minsize(window_width, window_height)
  Dispenser(root)
  root.mainloop()
```
